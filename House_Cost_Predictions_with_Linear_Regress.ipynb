#### *MSZoning*

```python
fig = px.histogram(train_prices_df, x='MSZoning', title='MSZoning: Distribution of Zoning Types', color='MSZoning')
fig.update_layout(bargap=0.1)
fig.show()
```

- From the plot above, we see Residential Low Density having the highest count of zoning types, followed by Residential Medium Density. 

#### *LotArea*

```python
fig2 = px.histogram(train_prices_df, x='LotArea', marginal='box', title='Lot Area Distribution')
fig2.update_layout(bargap=0.1)
fig2.show()
```

```python
fig = px.scatter(train_prices_df, 
                 x='SalePrice', 
                 y='LotArea',  
                 opacity=0.8, 
                 title='Prices vs. Lot Area')
fig.update_traces(marker_size=5)
fig.show()
```

```python
train_prices_df.SalePrice.corr(train_prices_df.LotArea)
```

- We can see that the largest count of houses per squarefoot area is from 8 to 9.9 thousand sqr ft. 
 - The median sqr footage is 9,478 sqr ft, the lowest being 1,300 and the largest (being an outlier in the data) being over 215k sqr ft.
 - there is very small correlation, if any, between the SalePrice and the LotArea


#### *BldgType*

```python
fig3 = px.histogram(train_prices_df, x='BldgType', color='BldgType', title='Housing Types')
fig3.update_layout(bargap=0.1)
fig3.show()
```

```python
fig4 = px.violin(train_prices_df, 
                 x='BldgType', 
                 y='SalePrice', 
                 color='BldgType',  
                 title='Price vs. Building Types')
fig4.update_traces(marker_size=5)
fig4.show()
```

#### *OverallCond*

```python
fig5 = px.histogram(train_prices_df, x='OverallCond', marginal='box', title='Housing Condition')
fig5.update_layout(bargap=0.1)
fig5.show()
```

```python
fig6 = px.strip(train_prices_df, 
                 x='OverallCond', 
                 y='SalePrice',  
                 title='Prices vs. Condition')
fig6.show()
```
