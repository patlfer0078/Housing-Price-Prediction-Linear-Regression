### Identify Numeric and Categorical Data

The next step in data preparation is to identify numeric and categorical columns. We can do this by looking at the data type of each column.

```python
prices_df.info()
```

>Now I create two lists named `numeric_cols` and `categorical_cols` containing names of numeric and categorical input columns within the dataframe. Numeric columns have data types  either `int64` or `float64`, while categorical columns have the data type `object`.
>
>[StackOverflow Finding Numeric Columns](https://stackoverflow.com/questions/25039626/how-do-i-find-numeric-columns-in-pandas).

```python
import numpy as np

numeric_cols = inputs_df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = inputs_df.select_dtypes(include=['object']).columns.tolist()
```

```python
print(list(numeric_cols))
```

```python
print(list(categorical_cols))
```

### Dealing with missing values in Numerical Data

Some of the numeric columns in our dataset contain missing values (`nan`).

```python
missing_counts = inputs_df[numeric_cols].isna().sum().sort_values(ascending=False)
missing_counts[missing_counts > 0]
```

ML models can't work with missing data so we have to fill any missing values through a process called [imputation](https://scikit-learn.org/stable/modules/impute.html).

There are many methods of imputation, but I went with the basic method, the `SimpleImputer` class from `sklearn.impute`.


```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'mean') # Create the imputer
imputer.fit(inputs_df[numeric_cols]) # Fit the imputer to the numeric columns
```

```python
list(imputer.statistics_)
```

After imputation, none of the numeric columns should contain any missing values.

```python
inputs_df[numeric_cols] = imputer.transform(inputs_df[numeric_cols]) # Transform and replace the numeric columns
missing_counts = inputs_df[numeric_cols].isna().sum().sort_values(ascending=False)
missing_counts[missing_counts > 0] # should be an empty list
```

### Scale Numerical Values

>The numeric columns in this dataset have varying ranges. A good practice is to [scale numeric features](https://scikit-learn.org/stable/modules/preprocessing.html#scaling-features-to-a-range) to a small range of values e.g. $(0,1)$.
>
>This feature ensures that no particular feature has a disproportionate impact on the model's loss. Optimization algorithms also work better in practice with smaller numbers.

```python
inputs_df[numeric_cols].describe().loc[['min', 'max']] # Identifying the max and min values for each column
```

> Here I scale the numeric values to the $(0, 1)$ range using `MinMaxScaler` from `sklearn.preprocessing`.


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Create the scaler
scaler = MinMaxScaler()
# Fit the scaler to the numeric columns
scaler.fit(inputs_df[numeric_cols])
```

```python
# Transform and replace the numeric columns
inputs_df[numeric_cols] = scaler.transform(inputs_df[numeric_cols])
inputs_df[numeric_cols].describe().loc[['min', 'max']] #ranges should be (0,1)
```

### Encode Categorical Columns

The dataset contains several categorical columns, each with a different number of categories. Since ML models can only be trained with numeric data, we need to convert categorical data to numbers. A common technique is to use one-hot encoding for categorical columns.

One hot encoding involves adding a new binary (0/1) column for each unique category of a categorical column.

```python
inputs_df[categorical_cols].nunique().sort_values(ascending=False)
```

> Now we encode the categorical columns in the dataset as one-hot vectors using `OneHotEncoder` from `sklearn.preprocessing`. Add a new binary (0/1) column for each category


```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#Create the encoder
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
#Fit the encoder to the categorical columns
encoder.fit(inputs_df[categorical_cols])
```

```python
# Generate column names for each category
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
encoded_cols
```

```python
len(encoded_cols)
```

```python
# perform transform and add new one-hot category columns
inputs_df[encoded_cols] = encoder.transform(inputs_df[categorical_cols])
```

The new one-hot category columns should now be added to `inputs_df`.

```python
inputs_df
```

### Training and Validation Set

Finally, split the dataset into a training and validation set. I chose to randomly select 25% subset of the data for validation. Also, I used just the numeric and encoded columns, since the inputs to the model must be numbers.

[test_train_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)

```python
from sklearn.model_selection import train_test_split
#assign the training and validation inputs and targets
#use the train_test_split function and test_size to 0.25
train_inputs, val_inputs, train_targets, val_targets = train_test_split(inputs_df[numeric_cols + encoded_cols],
                                                                        targets,
                                                                        test_size=0.25,
                                                                        random_state=42)
```

```python
train_inputs
```

```python
train_targets
```

```python
val_inputs
```

```python
val_targets
```

## Step 3 - Train a Linear Regression Model

Linear regression is a commonly used technique for solving [regression problems](https://www.geeksforgeeks.org/regression-classification-supervised-machine-learning/).The predictions from the model are evaluated using a loss function like the Root Mean Squared Error (RMSE).
>
>However, linear regression doesn't generalize very well when we have a large number of input columns with co-linearity i.e. when the values one column are highly correlated with values in other column(s). This is because it tries to fit the training data perfectly.

For this reason, I use Ridge Regression, a variant of linear regression that uses a technique called L2 regularization to introduce another loss term that forces the model to generalize better. https://www.youtube.com/watch?v=Q81RR3yKn30

```python
from sklearn.linear_model import Ridge
# to learn more about the Ridge function, type ?Ridge into an empty cell and run
```

```python
# Create the model
model = Ridge(solver='auto')
# Fit the model using inputs and targets
model.fit(train_inputs[numeric_cols + encoded_cols], train_targets)
```

```python
print(model.intercept_)
```

`model.fit` uses the following strategy for training the model:

1. Initialize a model with random parameters (weights & biases).
2. Pass some inputs into the model to obtain predictions.
3. Compare the model's predictions with the actual targets using the loss function.
4. We use an optimization technique (like least squares, gradient descent etc.) to reduce the loss by adjusting the weights & biases
5. Repeat steps 1 to 4 till the predictions from the model are good enough.



## Step 4 - Make Predictions and Evaluate the Model
Now that the model is trained, it can be used to generate predictions for the training and validation inputs. Evaluation of the model's performance can be used with the RMSE (root mean squared error) loss function.

```python
from sklearn.metrics import mean_squared_error
# make predictions for training inputs
train_preds = model.predict(train_inputs[numeric_cols + encoded_cols])
train_preds
```

```python
train_rmse = mean_squared_error(train_targets, train_preds, squared=False)
print('The RMSE loss for the training set is {}.'.format(train_rmse))
```

```python
#make predictions for validation input set
val_preds = model.predict(val_inputs[numeric_cols + encoded_cols])
val_preds
```

```python
val_rmse = mean_squared_error(val_targets, val_preds, squared=False)
```

```python
print('The RMSE loss for the validation set is {}.'.format(val_rmse))
```

### Making Predictions

The model can be used to make predictions on new inputs using the following helper function:

```python
def predict_input(single_input):
    input_df = pd.DataFrame([single_input])
    input_df[numeric_cols] = imputer.transform(input_df[numeric_cols])
    input_df[numeric_cols] = scaler.transform(input_df[numeric_cols])
    input_df[encoded_cols] = encoder.transform(input_df[categorical_cols].values)
    X_input = input_df[numeric_cols + encoded_cols]
    return model.predict(X_input)[0]
```
